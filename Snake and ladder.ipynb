```python
import random
snake_position={28:10,39:17,50:5,63:18,71:25,84:58,91:64,95:56,98:40}
ladder_position={2:23,6:36,9:30,20:59,34:72,45:96,53:75,65:99,74:92}

p1_current_position = 0
p2_current_position = 0

p1=input("enter player1 name: ")
p2=input("enter player2 name: ")
while True:
    input("\n""-> "+ p1 +","+ " press enter to roll the die......")
    dice_value=random.randint(1,6)
    print(" ","dice value is:",dice_value)
    old_position= p1_current_position
    p1_current_position = p1_current_position + dice_value  
    
    if (p1_current_position > 100):
        print("\n""you need",100-old_position,"to win")
        p1_current_position = old_position
        
    elif p1_current_position in snake_position:
        p1_current_position=snake_position[p1_current_position]
        print("\n""       OOPS!")
        print("\n""~~~~~~~SNAKE BITE~~~~~~~")
        
    elif p1_current_position in ladder_position:
        p1_current_position=ladder_position[p1_current_position]
        print("\n""       WOW!")
        print("\n""#######LADDER CLIMB#######")
        
    print("\n",p1,"current position is:",p1_current_position)
        
    if (p1_current_position == 100):
        print("\n","congratulations!", p1 ,"has won the game.")
        break  
    
    
    input("\n""-> "+ p2 +","+ " press enter to roll the die......")
    dice_value=random.randint(1,6)
    print(" ","dice value is:",dice_value)
    old_position= p2_current_position
    p2_current_position = p2_current_position + dice_value
    
    if (p2_current_position > 100):
        print("\n""you need",100-old_position,"to win")
        p2_current_position = old_position
        
        
    elif p2_current_position in snake_position:
        p2_current_position=snake_position[p2_current_position]
        print("\n""       OOPS!")
        print("\n""~~~~~~~SNAKE BITE~~~~~~~")
        
    elif p2_current_position in ladder_position:
        p2_current_position=ladder_position[p2_current_position]
        print("\n""       WOW!")
        print("\n""#######LADDER CLIMB#######")
              
    print("\n",p2,"current position is:",p2_current_position)
    
    if (p2_current_position == 100):
        print("\n""congratulations!", p2 ,"has won the game.")
        break
```
